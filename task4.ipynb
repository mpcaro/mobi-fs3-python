**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical $3 \times 3$ Prewitt filters:

```python
def prewitt_v(img):
    result = numpy.zeros(img.shape)
    filter = numpy.array([[-1,-1,-1], [0,0,0], [1,1,1]])
    for p in numpy.ndindex(img.shape[0]-2,img.shape[1]-2):
        neighborhood_sum = 0
        for q in numpy.ndindex(3,3):
            weight = filter[q[0],q[1]]
            y = p[0] + q[0]
            x = p[1] + q[1]
            varint = img[y,x] * weight
            neighborhood_sum += varint
        neighborhood_mean = neighborhood_sum / 6
        result[p[0]+1, p[1]+1] = neighborhood_mean
    return result
```

```python
def prewitt_h(img):
    result = numpy.zeros(img.shape)
    filter = numpy.array([[-1,0,1], [-1,0,1], [-1,0,1]])
    for p in numpy.ndindex(img.shape[0]-2,img.shape[1]-2):
        neighborhood_sum = 0
        for q in numpy.ndindex(3,3):
            weight = filter[q[0],q[1]]
            y = p[0] + q[0]
            x = p[1] + q[1]
            varint = img[y,x] * weight
            neighborhood_sum += varint
        neighborhood_mean = neighborhood_sum / 6
        result[p[0]+1, p[1]+1] = neighborhood_mean
    return result
```

```python
lena.copy()
meinefilterlena4 = prewitt_h(lena)
plt.imshow(meinefilterlena4)
```
